```python
with open('test.txt') as files:
    output = {}
    for line in files.readlines():
        line = line.strip()
        if line.count(',') == 1:
            output[line] = []
            currentPoint = line
        else:
            temp = list(map(float, line.split(',')))
            output[currentPoint].append(temp)
```
